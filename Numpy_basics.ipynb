# Dot product  / inner product

```python
a = np.array([1,2])
b = np.array([3,4])
```

```python
# manual way
dot = 0
for e,f in zip(a,b):
    dot +=e*f
print(dot)
```

### Sum over axis
innumpy axis =1 mean sum over each row

```python
m = np.array([[1,2,3],[3,2,1],[4,5,6]])
m
```

```python
# sum of row for each row
m.sum(axis=1,keepdims=True)
```

```python
# Sum over each column
print(m.sum(axis=0,keepdims=True).shape)
m.sum(axis=0,keepdims=True)
```

```python
# what is no keepdims=True? to keep the dimensions
print(m.sum(axis=0).shape)
m.sum(axis=0)
```

```python
row_sum = m.sum(axis=1,keepdims=True)
row_sum
```

```python
# Change the diagonal value of matrix by adding  log of row sum
t_matrix = np.copy(m)
import math
for i in range(len(m)):
    t_matrix[i,i] = t_matrix[i,i]+math.log(row_sum[i])
t_matrix
```

```python
# Same this can be achieved by vectorising
d= np.diag(m)
d.shape
```

```python
# So need to reshape
d = np.reshape(d,(3,1))
d.shape
```

```python
#perform vectorise function
d =d + np.vectorize(math.log)(row_sum)
d
```

```python
t_matrix2 = np.copy(m)
np.fill_diagonal(t_matrix2,d)
t_matrix2
```

```python
print(t_matrix2==t_matrix)
```

### Matrice multiplication manual vs numpy

```python
a = [[1,2,3],[2,3,4]]
b = [[2,1],[1,2],[3,1]]
ans = []
k=[]
for i in range(len(a)):
    for ii in range(len(b[i])):
        v=0
        for j in range(len(a[i])):
            aij = a[i][j]
            bji = b[j][ii]
            v += aij*bji
        k.append(v)
    ans.append(k)
    k=[]
            
```

```python
np.array([ans])
```
